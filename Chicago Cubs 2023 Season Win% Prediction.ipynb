```python
import pandas as pd
import numpy as np

# read in the data
df = pd.read_csv(r'C:\Users\trevj\Downloads\Chicago Cubs NL Central Stats.csv')

#we will be using data from when the cubs first entered the NL Central division
df
```

```python
#Finding the Cubs average wins over the last 29 seasons
df.loc[:, "W"].mean()
```

```python
#Finding the Cubs average losses over the last 29 seasons
df.loc[:, "L"].mean()
```

```python
#the problem with this data is that is includes seasons where they did not play a full 162 games
```

```python
#lets throw out any season where they did not play at least 160 games
```

```python
#adding new column for total games
df.loc[:,"W" : "L"]
df['Total Games'] = df['W'] + df['L']
```

```python
df
```

```python
#we can see that there are 3 seasons where they did not play at least 160 games in a season
#lets throw those out so we can find a more accurate average based on a full season
```

```python
#dropping rows that have less than 160 games played
df.drop(df[df['Total Games'] <= 160].index, inplace = True)
```

```python
df
```

```python
#the new average wins
df.loc[:, "W"].mean()
```

```python
#new average losses
df.loc[:, "L"].mean()
```

```python
#average wins went from 76.75862068965517 to 79.61538461538461
#average losses went from 79.37931034482759 to 82.34615384615384
#the win/loss percentage stayed fairly consistant
```

```python
#add win percantage column
df['Win%'] = df['W'] / df['Total Games']
```

```python
df
```

```python
#lets look at the cubs trend win/loss trend on a chart
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
plt.scatter(x=df['Year'],y=df['Win%'])
plt.title('Wins vs Losses')
plt.xlabel('Year')
plt.ylabel('Win%')
plt.show()
```

```python
sns.regplot(x='Year',y='Win%',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"});
```
